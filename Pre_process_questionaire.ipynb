```python
#This creates files from Gorilla

import pandas as pd
import os
```

```python
in_path = f"{os.getcwd()}\\Data\\Analysis\\Questionnaire_Data\\"
files = os.listdir(in_path)
print(files)
```

```python
out_path = f"{os.getcwd()}\\Data\\Analysis\\"
```

```python
bg = pd.read_csv(f"{in_path}{files[0]}")
bg.columns
```

```python
clean_columns = ['Participant Private ID', 'Participant Status', 'Task Name', 'Page ID', 'Question','Response','Object ID' ]
bg_clean = bg[clean_columns]
bg_clean
```

```python
# Pivot table
bg_compiled = bg_clean.pivot_table(
    index='Participant Private ID',  # rows
    columns='Question',              # columns
    values='Response',               # values to fill
    aggfunc='last'  
).reset_index()
```

```python
bg_compiled.head()
bg_compiled.to_csv(f"{out_path}Participant_Background.csv")
```

## Ai task

```python
ai_task = pd.read_csv(f"{in_path}{files[1]}")
ai_task.columns
```

```python
ai_task_columns = ['Participant Private ID','Page', 'Question', 'Response', 'Response Type']
ai_task_clean = ai_task[ai_task_columns]
ai_task_clean
```

```python
ai_task_clean = ai_task_clean[ai_task_clean['Response Type'] == 'response']
ai_task_clean = ai_task_clean.drop_duplicates(
    subset=["Participant Private ID", "Page", "Question"], 
    keep="last"
)

ai_task_columns = ['Participant Private ID','Page', 'Question', 'Response']
ai_task_clean = ai_task_clean[ai_task_columns]

ai_task_clean
```

```python
ai_task_clean.to_csv(f"{out_path}Clean_Task.csv")
```
